```python
import numpy as np
def calculate_tolerance_accuracy(actual_salaries, predicted_salaries, tolerance_percentage):
    correct_predictions = 0
    tolerance_factor = 1 + (tolerance_percentage / 100)

    for actual, predicted in zip(actual_salaries, predicted_salaries):
        lower_bound = actual / tolerance_factor
        upper_bound = actual * tolerance_factor

        if lower_bound <= predicted <= upper_bound:
            correct_predictions += 1

    accuracy_within_tolerance = correct_predictions / len(actual_salaries)
    return accuracy_within_tolerance

# Example usage
#salary label is actural_salaries and predicted_salaries is linear regression prediction
#convert y_test to numpy array
actual_salaries = y_test
predicted_salaries = ans
tolerance_percentage = 10  # 10% tolerance

accuracy = calculate_tolerance_accuracy(actual_salaries, predicted_salaries, tolerance_percentage)
print(f'Accuracy within tolerance: {accuracy * 100:.2f}%')
```

```python
import numpy as np

def calculate_accuracy_within_range_part2(actual_salaries, predicted_salaries, tolerance=0.20):
    """
    Calculate the accuracy of predictions within a certain tolerance percentage of the actual values.

    :param actual_salaries: Array of actual salary values.
    :param predicted_salaries: Array of predicted salary values.
    :param tolerance: Tolerance percentage for considering a prediction as accurate.
    :return: The accuracy of predictions within the specified tolerance.
    """
    # Calculate the acceptable range (lower and upper bounds) for each actual salary
    lower_bounds = actual_salaries * (1 - tolerance)
    upper_bounds = actual_salaries * (1 + tolerance)

    # Determine if each prediction is within the acceptable range
    correct_predictions = np.logical_and(predicted_salaries >= lower_bounds, predicted_salaries <= upper_bounds)

    # Calculate the accuracy as the proportion of correct predictions
    accuracy = np.mean(correct_predictions)

    return accuracy
```
